## Python Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Cleaning

### Import datasets

```python
# Import the raw dataframe from assets folder
df_train = pd.read_csv('../assets/train.csv')
df_test = pd.read_csv('../assets/test.csv')
```

```python
# Train df details 
print(df_train.shape)
df_train.info()
df_train.head()
```

```python
# Test df details 
print(df_test.shape)
df_test.info()
df_test.head()
```

Datatypes for all the columns except `Date` seems fine. We will proceed to update the datatype for `Date` to datetime64 dtype.

```python
# Convert Date to Datetime type (train df)
df_train['Date']=pd.to_datetime(df_train['Date'], format='%Y-%m-%d')

# Convert Date to Datetime type (test df)
df_test['Date']=pd.to_datetime(df_test['Date'], format='%Y-%m-%d')
```

## Exploratory Data Analysis - `train`

```python
# Check for missing values
df_train.isnull().sum()
```

* Based on Data Description, we can infer that there should be no duplicated observations. This is because the dataset is organized in a way that when the number of mosquitos exceed 50, they are split into another record (another row in the dataset), such that the number of mosquitos are capped at 50. Any remainder count of mosquitoes beyond counts of 50 will be assigned to a new row by itself.
* There are no missing values in `train` dataset.

### `Date` variable

```python
# Break down the Dates into Year & Month
df_train['year'] = df_train['Date'].dt.year
df_train['month'] = df_train['Date'].dt.month
```

```python
# List of unique months in train df
list(df_train['month'].unique())
```

This is aligned to the months of data collection, as described in Data Description.

### Understanding `AddressAccuracy`

Based on Data Dictionary, observations with low address accuracy would mean that Geocoder is not able to provide an accurate approximation of the closest address possible to the trap. In turn, the latitude and longitude captured might not be accurate to the location of the traps. 

```python
# Unique values of Address Accuracy
df_train['AddressAccuracy'].value_counts(normalize=True)
```

Since about 0.86% of records have a low address accuracy of 3, the deviation in location accuracy in the traps would not be significant. Thus, we can assume that the latitude and longitude in the entire dataset is accurate and useful for EDA and modelling later.

```python
# Bar chart for observations count based on address accuracy
ax = sns.countplot(data=df_train,
                   x=df_train['AddressAccuracy']);

plt.title('No. of observations based on AddressAccuracy')
ax.bar_label(ax.containers[0]);
```

### Location of Traps

```python
# Count of Mosquito Traps in train df
print(f"There are {df_train['Trap'].nunique()} distinct traps recorded in train dataset.")
```

```python
# Create df for all coordinates of traps
df_coordinate = df_train[['Latitude','Longitude']]

# Create df for all coordinates of traps, provied WNV is present
df_coor_wnv = df_train[df_train.WnvPresent == 1][['Latitude','Longitude']]
```

```python
# prepare the bounding box based on min-max coordinates
min_max_coor = (df_coordinate.Longitude.min(), df_coordinate.Longitude.max(), df_coordinate.Latitude.min(), df_coordinate.Latitude.max())
print(min_max_coor)
```

```python
# Set some leeway for the bounding box
BBox = (-88.0, -87.5, 41.6, 42.1)
```

Using OpenStreetMap, we will plot the coordinates of the traps as scatter points on the map of Chicago.

```python
# Load map of Chicago image
map_ca = plt.imread('../images/chicago_map.png')

plt.figure(figsize = (9,10))

# Plot coordinates of traps overall
plt.scatter(x=df_coordinate.Longitude,
            y=df_coordinate.Latitude,
            zorder=1,
            c='b',
            s=20
           )

# Plot coordinates of traps with WNV present
plt.scatter(x=df_coor_wnv.Longitude,
            y=df_coor_wnv.Latitude,
            zorder=2,
            alpha=0.5,
            c='r',
            s=10,
            marker='x'
           )

plt.title('Location of Traps with WNV present across Chicago (Train)')
plt.xlim(BBox[0],BBox[1])
plt.ylim(BBox[2],BBox[3])
plt.legend(['Traps','Traps with WNV'])

# Incorporate the map into the scatter plot
plt.imshow(map_ca, zorder=0, extent = BBox, aspect= 'equal');
```

### Comparison between observations with WNV and without WNV

```python
# Bar chart for count of records with WNV & without WNV
ax = sns.countplot(data=df_train,
                   x=df_train['WnvPresent']);

plt.title('No. of observations by WnvPresent variable')
ax.bar_label(ax.containers[0]);
```

```python
# Proportion of WNV present observations against WNV absense observations
df_train['WnvPresent'].value_counts(normalize=True)
```

The proportion of WNV Present records against WNV absent records is very unbalanced. Only 5% of the total records in train dataset are WNV present. Additional techniques will be needed to handle the imbalanced classes prior to model training and avoid the metric trap.

### Mosquito Species

```python
# Create a df for Species & Mosquito count, provided that there is a presence of WNV
mos_species_wnv = df_train[df_train['WnvPresent']==1][['Species', 'NumMosquitos']]
```

```python
# Count plot of Mosquitoes caught by species
total_mos = df_train[['Species', 'NumMosquitos']].groupby(by='Species').sum().sort_values('NumMosquitos')

ax = total_mos.plot(kind='bar', legend=False)

plt.xticks(rotation=90)
plt.title(f'Total overall count of mosquitos caught by Species')
ax.bar_label(ax.containers[0]);
```

```python
# Count plot of Mosquitoes caught with WNV by species
total_virus = mos_species_wnv.groupby(by='Species').sum().sort_values('NumMosquitos')

ax = total_virus.plot(kind='bar', legend=False)

plt.xticks(rotation=0)
plt.title(f'Total count of mosquitos caught with WNV by Species')
ax.bar_label(ax.containers[0]);
```

So far from the `train` dataset, only species under `CULEX PIPIENS` and `CULEX RESTUANS` have the presense of WNV. However, from research studies, other species under the Culex family are main vectors of transmission across the U.S. ([source](https://www.vdci.net/vector-borne-diseases/west-nile-virus-education-and-mosquito-management-to-protect-public-health/)). It is good to infer that `CULEX PIPIENS` and `CULEX RESTUANS` are the main vectors in Chicago specfically is due to their high abundance in the city.

### Mosquitoes caught by Year/Month trend

```python
# Line chart of mosquitoes caught overall
sns.set_style('whitegrid')

ax = sns.lineplot(x='month',
                  y='NumMosquitos', 
                  hue='year', 
                  data=df_train[['month', 'year', 'NumMosquitos']].groupby(by=['month','year'], as_index=False).sum(),
                  palette="tab10"
                 )

plt.ylabel('Mosquito Count')
plt.title(f'Total count of mosquitoes caught by month on month');
```

```python
# Create a df for Mosquito count across month/year, provided that there is a presence of WNV
mos_mthyr_wnv = df_train[df_train['WnvPresent']==1][['month','year','NumMosquitos']]
```

```python
# Line chart of mosquitoes caught with WNV
sns.set_style('whitegrid')

ax = sns.lineplot(x='month',
                  y='NumMosquitos', 
                  hue='year', 
                  data=mos_mthyr_wnv.groupby(by=['month','year'], as_index=False).sum(),
                  palette="tab10"
                 )

plt.ylabel('Mosquito Count')
plt.title(f'Total count of mosquitoes caught with WNV by month on month');
```

The occurance of WNV usually starts from July to September where the peak lies within August and September. In 2013, the occurance of WNV spiked drastically, as compared to 2011. This could be due to a rise in number of mosquitoes caught for 2013.

## EDA Summary on `train` & `test` datasets

Most of the traps are located mainly in the metropolitan area of Chicago where the residential houses are situated. There is a higher concentration of traps located in the residential areas of the city, as compared to non-residential areas (i.e. the airport). In total, there are 136 distinct traps recorded in the train dataset.

WNV presence usually peaks in August, which coincides with that of the mosquitoes population caught by the traps. The species genus `CULEX PIPIENS` and `CULEX RESTUANS` has been found to be major vectors of transmission in Chicago. This can be inferred that these 2 species are the dominant species in the city and thus stands a higher likelihood of being transmission vectors for WNV.

The `test` dataset is about 11 times larger than the `train` dataset in terms of observations. In `test` dataset, there are 13 more distinct traps recorded, as compared to that of `train` dataset. Fortunately, the different traps are not located far away from existing traps stated in `train` dataset. We can use clustering methods to cluster them together for modeling later. In terms of mosquito species, `test` dataset has an additional class `UNSPECIFIED CULEX` that does not exists in `train` dataset. This will need to manage exceptionally during One-Hot Encoding.
